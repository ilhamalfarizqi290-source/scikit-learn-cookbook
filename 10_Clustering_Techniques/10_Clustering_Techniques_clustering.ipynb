# Chapter 10 – Clustering Techniques

## Objective
Understanding unsupervised learning through clustering.

## Theoretical Background

Clustering groups similar data points without labels.

K-Means algorithm:
- Chooses k centroids
- Assigns points to nearest centroid
- Updates centroids iteratively

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

X, _ = load_iris(return_X_y=True)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

kmeans.cluster_centers_
```

## Analysis

K-Means discovers structure in unlabeled data.

Cluster centers represent typical patterns in each group.
