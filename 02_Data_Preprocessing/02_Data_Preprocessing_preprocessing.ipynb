# Chapter 2 – Data Preprocessing

This chapter demonstrates essential preprocessing techniques before training machine learning models.

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

# Example dataset with missing values
data = pd.DataFrame({
    "Feature1": [1, 2, np.nan, 4],
    "Feature2": [np.nan, 3, 4, 5]
})

imputer = SimpleImputer(strategy="mean")
imputed_data = imputer.fit_transform(data)

pd.DataFrame(imputed_data, columns=data.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

categorical_data = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Blue"]
})

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(categorical_data)

pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(imputed_data)

pd.DataFrame(scaled, columns=data.columns)
```

## Analysis

Data preprocessing improves model performance by:
- Handling missing values
- Converting categorical features into numerical form
- Normalizing feature scales

Proper preprocessing ensures stable and efficient optimization.
