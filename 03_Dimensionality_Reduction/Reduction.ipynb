# Chapter 3 – Dimensionality Reduction

## Objective
This chapter explores dimensionality reduction techniques to simplify high-dimensional datasets while preserving important information.

## Theoretical Background

High-dimensional data may cause:

- Increased computational cost
- Overfitting
- Curse of dimensionality

Dimensionality reduction aims to:
- Reduce feature space
- Preserve meaningful variance
- Improve model generalization

Principal Component Analysis (PCA) transforms the original features into a smaller number of orthogonal components that capture maximum variance.

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd

# Load dataset
X, y = load_iris(return_X_y=True)

# Apply PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

pd.DataFrame(X_reduced, columns=["PC1", "PC2"]).head()
```

## Analysis

The dataset originally contains 4 features.
After applying PCA with 2 components, the dimensionality is reduced to 2 principal components.

These components capture most of the variance in the data, allowing:
- Simpler visualization
- Reduced computational complexity
- Potential improvement in generalization
