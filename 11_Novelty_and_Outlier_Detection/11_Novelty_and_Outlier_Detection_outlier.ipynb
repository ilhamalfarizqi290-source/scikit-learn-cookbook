# Chapter 11 – Novelty and Outlier Detection

## Objective
Detecting anomalies and rare observations.

## Theoretical Background

Outliers are observations that deviate significantly from others.

Isolation Forest isolates anomalies using random partitioning.

```python
from sklearn.ensemble import IsolationForest
import numpy as np

rng = np.random.RandomState(42)
X = 0.3 * rng.randn(100, 2)

clf = IsolationForest(contamination=0.1, random_state=42)
clf.fit(X)

clf.predict(X)
```

## Analysis

Isolation Forest identifies unusual data points efficiently.

Outlier detection improves robustness of machine learning systems.
