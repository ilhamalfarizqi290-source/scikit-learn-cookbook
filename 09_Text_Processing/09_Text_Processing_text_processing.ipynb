# Chapter 9 – Text Processing

## Objective
Transforming textual data into numerical features.

## Theoretical Background

Machine learning models require numerical input.

Text must be converted using:
- Bag of Words
- TF-IDF (Term Frequency–Inverse Document Frequency)

TF-IDF reduces importance of common words.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "machine learning is powerful",
    "scikit learn makes machine learning easy",
    "text data must be converted to numbers"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

X.toarray()
```

## Analysis

TF-IDF converts raw text into weighted numerical vectors.

This allows machine learning models to process textual information.
