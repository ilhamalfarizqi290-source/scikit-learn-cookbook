# Chapter 13 – Deploying Models

## Objective
Understanding how trained models are deployed in real applications.

## Theoretical Background

Deployment involves:
- Saving trained model
- Loading model in production
- Monitoring performance

Model serialization allows reuse without retraining.

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import joblib

X, y = load_iris(return_X_y=True)

model = LogisticRegression(max_iter=200)
model.fit(X, y)

joblib.dump(model, "model.pkl")
```

## Analysis

The trained model is saved to disk and can be loaded later
for predictions in production environments.
